###Process decimals of Latitude and Longitude 
The resort data's latitude (lat) and longitude (long) are random numbers, and those from the snow data are ending with 0.125, 0.375, 0.625, 0.875. In order to match two of them, round the lat and long in resort data.


```python
snow_data.Latitude.diff().unique()
```

```python
# Define a function to round input number to specific multiple.
def round_to_multiple(number, multiple):
    result = multiple * round(number / multiple)
    return result
    
#Store the rounded data in 'r_lat' and 'r_long' columns.
resort_data['r_lat'] = round_to_multiple(resort_data['Latitude'],0.125)
resort_data['r_long'] = round_to_multiple(resort_data['Longitude'],0.125)
resort_data[['r_lat','r_long']]
```

```python
# Deal with those rounded lat and long ending with .250, .500, .750 and .000.

resort_data.loc[ (resort_data['r_lat']%0.250 == 0 )& (resort_data['r_lat']>resort_data['Latitude']),'r_lat']=resort_data['r_lat']-0.125
resort_data.loc[ (resort_data['r_lat']%0.250 == 0 )& (resort_data['r_lat']<=resort_data['Latitude']),'r_lat']=resort_data['r_lat']+0.125

resort_data.loc[ (resort_data['r_long']%0.250 == 0 )& (resort_data['r_long']>resort_data['Longitude']),'r_long']=resort_data['r_long']-0.125
resort_data.loc[ (resort_data['r_long']%0.250 == 0 )& (resort_data['r_long']<=resort_data['Longitude']),'r_long']=resort_data['r_long']+0.125

resort_data[['Latitude','r_lat','Longitude','r_long']].sort_values(by = 'Latitude')
```

```python
# Create a new column 'lat_long' to store the pair of latitude and longtitude columns in both datasets.
snow_data['lat_long'] = tuple(zip(snow_data.Latitude,snow_data.Longitude))
resort_data['lat_long'] = tuple(zip(resort_data.r_lat,resort_data.r_long))
```

```python
snow_data.head()
```

```python
resort_data.head()
```

```python
#Check if there is any row in snow data that ends with unexpected decimals.
snow_data['r_lat']= round_to_multiple(snow_data['Latitude'],0.125)
snow_data['r_long'] = round_to_multiple(snow_data['Longitude'],0.125)
snow_data.query('r_lat!=Latitude or r_long!= Longitude')
```

### Groupby (Lat, Long), Max Snow data
Add a new column for maximum value of snow coverage for each pair of lat and long.

```python
snow_query = snow_data.groupby(['lat_long'], as_index=False).agg({'Snow':'max', 'Month':'first','lat_long':'first'})
snow_query = snow_query.rename(columns={"Snow": "Max_snow"})
snow_query = snow_query.rename(columns={"Month": "Max_snow_month"})
snow_query
```

### Merge Resort data and Snow data

```python
# merge
resort_snow = pd.merge(left=resort_data, right=snow_query, how='left',left_on='lat_long', right_on='lat_long')
resort_snow.head()
```

```python
# check if there is null value in Snow or Month
resort_snow[['Latitude','r_lat','Longitude','r_long','Max_snow_month','Max_snow','lat_long']].isna().sum()
```

```python
resort_snow[['Latitude','r_lat','Longitude','r_long','Max_snow_month','Max_snow','lat_long']].query('Max_snow_month != Max_snow_month or Max_snow!=Max_snow')
```

```python
# drop the rows we just found with NaNs
drop_query = resort_snow.query('Max_snow_month != Max_snow_month or Max_snow!=Max_snow').index
resort_snow = resort_snow.drop(drop_query,axis=0)
```
